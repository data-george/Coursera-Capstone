Let's use this function I learned from my IBM class to find the 100 venues/locations surrounding a 500 meter radius from each neighborhood location, of course we can alter the parameters but I think this is sensible if we consider this is a one kilometer diameter what's surrounding the home and the home for now is the centroid of the neighborhood (average of latitudes and longitudes)

```python
LIMIT = 100
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# calling the function getNearbyVenues to create the dataframe

Chicago_Venues = getNearbyVenues(names=Chicago_Home_Locations['Neighborhood'],
                                   latitudes =Chicago_Home_Locations['Latitude'],
                                   longitudes =Chicago_Home_Locations['Longitude']
                                  )
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

### CLUSTERING

We'll import KMeans first to create the model

```python
from sklearn.cluster import KMeans
```

I'm not scaling the data since it is already based on the probability of the prediction and the sum of the probabilities is always equal to one. Therefore our data is ready aside from removing the non numerical column which we will do next:

```python
#from sklearn.preprocessing import StandardScaler

#SS = StandardScaler()
Chicago_Clustering = Chicago_grouped.drop('Neighborhood', 1)
#SS.fit(Chicago_Clustering)
#data_transformed = SS.transform(Chicago_Clustering)
data_transformed = Chicago_Clustering
data = data_transformed
datachanged = data
```

Next, we will use the following loop to run KMeans at multiple ranges of K plotted against the sum of the within clusters squared distance, we want to minimize K (minimize within clusters square distance and maximize between clusters square distance so that the centroids represent true averages of different groups). We will plot the data and will choose the K at which the curve breaks at the 'elbow' which will be the optimal value:

```python
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(datachanged)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(data_transformed)
    data_transformed["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for n_cluster in range(2, 11):
    kmeans = KMeans(n_clusters=n_cluster).fit(data)
    label = kmeans.labels_
    sil_coeff = silhouette_score(data_transformed, label, metric='euclidean')
    print("For n_clusters={}, The Silhouette Coefficient is {}".format(n_cluster, sil_coeff))
```

Looks like 3 to me!

```python
# set number of clusters
kclusters = 3

Chicago_Clustering = Chicago_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Chicago_Clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

### EXPLORING THE CLUSTERS

Let's used the centroids developed by the algorithm to dig deeper into how these clusters are divided:

```python
centers = kmeans.cluster_centers_
```

We'll use the following function to get the probability of each category within each cluster prediction:

```python
# Function that creates a DataFrame with a column for Cluster Number

def pd_centers(Chicago_Clustering, centers):
	colNames = list(Chicago_Clustering)
	colNames.append('prediction')

	# Zip with a column called 'prediction' (index)
	Z = [np.append(A, index) for index, A in enumerate(centers)]

	# Convert to pandas data frame for plotting
	P = pd.DataFrame(Z, columns=colNames)
	P['prediction'] = P['prediction'].astype(int)
	return P
```

```python
P = pd_centers(Chicago_Clustering, centers)
P
```
